```python
import geopandas as gpd
```

Here we'll plot the ratio of tests to number of people diagnosed.

```python
%matplotlib inline
vic_poly = gpd.read_file('http://langlynx.ddns.net:8000/geojson_poly_data?schema=admin_0')
vic_poly['test_total_ratio'] = vic_poly['total'] / vic_poly['tests_total']

vic_poly.plot(
    column="test_total_ratio",
    legend=True,
    scheme="quantiles",
    figsize=(15, 15),
    missing_kwds={
        "color": "lightgrey", "edgecolor": "red", "hatch": "///", "label": "Missing values",
    },
)
```

Next, we'll plot the number of tests performed in each country as points.

```python
%matplotlib inline
vic_points = gpd.read_file('http://langlynx.ddns.net:8000/geojson_point_data?schema=admin_0')
vic_points.plot(
    column="tests_total",
    legend=True,
    scheme="quantiles",
    figsize=(15, 10),
    missing_kwds={
        "color": "lightgrey",
        "edgecolor": "red",
        "hatch": "///",
        "label": "Missing values",
    },
)
```
