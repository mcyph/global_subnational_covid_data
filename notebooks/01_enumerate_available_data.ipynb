```python
import requests
from pprint import pprint
```

We'll see which kinds of schemas we can get.

Schemas allow getting different covid-19 statistics at geographic levels of
different granularity.

For instance, `admin_0` refers to country-level data,
and `admin_1` to first-level administrative level (states/provinces etc).

```python
requests.get('http://langlynx.ddns.net:8000/schemas').json()
```

Next, we'll see the possible kinds of datatypes.

```python
requests.get('http://langlynx.ddns.net:8000/datatypes').json()
```

Next, we'll see the possible kinds of sources.

```python
requests.get('http://langlynx.ddns.net:8000/source_ids').json()
```

We can then see what kinds of datasources and datatypes are available
for a given schema. For example, for Local Government Areas (LGAs) of Australia:

```python
sorted(requests.get('http://langlynx.ddns.net:8000/case_data_sources',
                    params={'schema': 'lga'}).json())
```

```python
sorted(requests.get('http://langlynx.ddns.net:8000/case_datatypes',
                    params={'schema': 'lga'}).json())
```

After that, we'll download the data itself. In this case, we'll just get the current
Victorian active data by LGA by using the endpoint `/latest_case_data`, but
`/case_data_time_series` can also retrieve all data available to date.

```python
pprint(requests.get('http://langlynx.ddns.net:8000/latest_case_data',
                    params={'schema': 'lga',
                            'region_parent': 'au-vic',
                            'datatype': 'total'}
                    ).json())
```
