Next, we'll import the needed modules. descartes and geopandas are used 

```python
import descartes
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

vic_poly = gpd.read_file('http://langlynx.ddns.net:8000/geojson_poly_data?'
                         'schema=lga&'
                         'region_parent=au-vic&'
                         'datatype=total')
vic_poly.plot()
```

```python
%matplotlib inline
vic_points = gpd.read_file('http://langlynx.ddns.net:8000/geojson_point_data?'
                           'schema=lga&'
                           'region_parent=au-vic&'
                           'datatype=total')
vic_points.plot()
```
